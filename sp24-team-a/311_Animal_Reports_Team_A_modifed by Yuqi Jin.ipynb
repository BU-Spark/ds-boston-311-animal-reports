```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/FullAnimalReports.csv')
df = df.rename(columns={'open_dt': "datetime"})
df["datetime"] = pd.to_datetime(df["datetime"])
df["month"] = df["datetime"].dt.month

fig, axs = plt.subplots(figsize=(12, 4))
df.groupby(df["datetime"].dt.year)['closure_reason'].count().plot( kind='bar', rot=0, ax=axs )
plt.xlabel("Per year")
plt.ylabel("count")

fig, axs = plt.subplots(figsize=(12, 4))
df.groupby(df["datetime"].dt.month)['closure_reason'].count().plot( kind='bar', rot=0, ax=axs )
plt.xlabel("Per month")
plt.ylabel("count")

fig, axs = plt.subplots(figsize=(12, 4))
df.groupby(df["datetime"].dt.hour)['closure_reason'].count().plot( kind='bar', rot=0, ax=axs )
plt.xlabel("Per hour")
plt.ylabel("count")
```

# Create mapping of reports for animals aggregated by zip code

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/FullAnimalReports.csv')

# Plotting the data
plt.figure(figsize=(12, 6))
df.groupby('location_zipcode')['case_title'].count().plot(kind='bar', color='skyblue')
plt.xlabel('Zip Code')
plt.ylabel('Number of Cases')
plt.title('Number of Cases by Zip Code')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
df = pd.read_csv('/content/drive/MyDrive/FullAnimalReports.csv')

# Convert 'open_dt' column to datetime format
df['open_dt'] = pd.to_datetime(df['open_dt'])

# Extract the hour component from the datetime
df['hour'] = df['open_dt'].dt.hour

# Group by hour and count the number of animal reports
hourly_counts = df.groupby('hour').size()

# Plotting the data as a line graph
plt.figure(figsize=(10, 6))
hourly_counts.plot(kind='line', marker='o', color='orange')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Animal Reports')
plt.title('Number of Animal Reports per Hour')
plt.xticks(hourly_counts.index)
plt.grid(True)
plt.tight_layout()
plt.show()
```
