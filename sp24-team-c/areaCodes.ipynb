```python
import matplotlib.pyplot as plt
import pandas as pd
import os

# Assuming df is your DataFrame with 'open_dt', 'longitude', and 'latitude' columns

# Convert 'open_dt' column to datetime type if it's not already
df['open_dt'] = pd.to_datetime(df['open_dt'])

# Create a directory to store the images
output_dir = 'plots'
os.makedirs(output_dir, exist_ok=True)

# Loop through each day from 2011 to 2012
start_date = pd.to_datetime('2011-7-01')
end_date = pd.to_datetime('2012-12-31')
for single_date in pd.date_range(start_date, end_date):
    # Filter DataFrame for the specific date
    df_specific_date = df[df['open_dt'].dt.date == single_date.date()]

    # Plot the points for the specific date
    plt.scatter(df_specific_date['longitude'], df_specific_date['latitude'])
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title(f'Map of Points - {single_date.strftime("%Y-%m-%d")}')
    
    # Save the plot as an image
    plt.savefig(f'{output_dir}/{single_date.strftime("%Y-%m-%d")}.png')
    plt.close()

# Convert images to a GIF
import imageio
images = []
for filename in sorted(os.listdir(output_dir)):
    if filename.endswith('.png'):
        images.append(imageio.imread(os.path.join(output_dir, filename)))
imageio.mimsave('map_animation.gif', images, fps=5)

# Remove the temporary image files
import shutil
shutil.rmtree(output_dir)

print("GIF created successfully.")
```

```python
df = pd.read_csv('2011_2024.csv')
df['open_dt'] = pd.to_datetime(df['open_dt'])
```

```python
import os
import time
from datetime import datetime
from selenium import webdriver
import imageio
from PIL import Image, ImageDraw, ImageFont
import folium
from folium.plugins import HeatMap
import pandas as pd
from bs4 import BeautifulSoup

def add_text_to_image(image_filename, text):
    # Load the image
    base_image = Image.open(image_filename)

    # Get the dimensions of the base image
    base_width, base_height = base_image.size

    # Create a new image with white background
    text_image = Image.new("RGB", (base_width, 100), "white")

    # Set up font properties
    font = ImageFont.truetype("/System/Library/Fonts/Supplemental/arial.ttf", 50)

    # Calculate text size and position
    draw = ImageDraw.Draw(text_image)
    text_bbox = draw.textbbox((0, 0), text, font=font)
    text_x = (base_width - text_bbox[2]) // 2
    text_y = (100 - text_bbox[3]) // 2

    # Draw the text on the image
    draw.text((text_x, text_y), text, fill="black", font=font)

    # Combine the two images
    combined_image = Image.new("RGB", (base_width, base_height + 100))
    combined_image.paste(base_image, (0, 0))
    combined_image.paste(text_image, (0, base_height))

    # Save the combined image
    combined_image.save(image_filename)

def modify_zoom_level(html_file):
    with open(html_file, 'r') as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the script tag containing the zoom level
    script_tag = soup.find('script', string=lambda text: 'zoom' in str(text))

    if script_tag:
        # Extracting the script content
        script_content = script_tag.string.strip()

        # Replacing the zoom level from 10 to 12
        script_content = script_content.replace('zoom: 10', 'zoom: 12')

        # Update the script content
        script_tag.string.replace_with(script_content)

        # Write the modified HTML back to the file
        with open(html_file, 'w') as file:
            file.write(str(soup))

# Directory containing the HTML files
html_dir = 'maps'

# Create a directory to store the HTML files
os.makedirs(html_dir, exist_ok=True)

# Load DataFrame with latitude and longitude columns
# Assuming df is your DataFrame with 'latitude' and 'longitude' columns
# Drop rows with NaN values in latitude or longitude columns
df_cleaned = df.dropna(subset=['latitude', 'longitude'])

# Loop through each year from 2013 to 2023
for year in range(2013, 2024):
    # Create a directory to store images for this year
    year_image_dir = os.path.join(html_dir, str(year))
    os.makedirs(year_image_dir, exist_ok=True)

    # Loop through each month
    for month in range(1, 13):
        # Filter DataFrame for the specific year and month
        df_specific_month = df_cleaned[(df_cleaned['open_dt'].dt.year == year) & (df_cleaned['open_dt'].dt.month == month)]

        # Create a map object
        m = folium.Map(location=[df_specific_month['latitude'].mean(), df_specific_month['longitude'].mean()], zoom_start=10)

        # Add heatmap layer to the map
        heat_data = [[row['latitude'], row['longitude']] for _, row in df_specific_month.iterrows()]
        HeatMap(heat_data).add_to(m)

        # Save the map as an HTML file
        file_name = os.path.join(year_image_dir, f'{year}-{month:02}.html')
        m.save(file_name)

        # Modify zoom level in the HTML file
        modify_zoom_level(file_name)

# Directory to save GIFs
gif_dir = 'gif'
os.makedirs(gif_dir, exist_ok=True)

# Loop through each year and create GIF
for year in range(2013, 2024):
    # Directory to save image files
    year_image_dir = os.path.join(html_dir, str(year))

    # List to store image paths
    image_paths = []

    # Configure Selenium webdriver
    options = webdriver.ChromeOptions()
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--headless')
    driver = webdriver.Chrome(options=options)

    # Loop through HTML files and capture screenshots
    for filename in sorted(os.listdir(year_image_dir)):
        if filename.endswith('.html'):
            html_file = os.path.abspath(os.path.join(year_image_dir, filename))

            # Open HTML file in headless Chrome and adjust zoom level
            driver.get(f'file://{html_file}')
            time.sleep(2)  # Wait for the page to load

            # Capture screenshot
            screenshot = driver.get_screenshot_as_png()

            # Save screenshot as PNG
            image_file = os.path.join(year_image_dir, filename.replace('.html', '.png'))
            with open(image_file, 'wb') as f:
                f.write(screenshot)

            # Add filename (without extension) as text to image
            file_text = os.path.splitext(filename)[0]  # Extract filename without extension
            add_text_to_image(image_file, file_text)

            image_paths.append(image_file)

    # Convert screenshots to GIF
    images = [imageio.imread(image_path) for image_path in image_paths]
    gif_filename = os.path.join(gif_dir, f'{year}.gif')
    imageio.mimsave(gif_filename, images, fps=1)

    # Remove temporary image files
    for image_path in image_paths:
        os.remove(image_path)

    # Quit the webdriver
    driver.quit()

    print(f"GIF for year {year} created successfully.")

print("All GIFs created successfully.")
```
